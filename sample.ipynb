```python
def get_list_columns_names(df: pd.DataFrame):
    columns_names = "["

    for idx, col in enumerate(df.columns):
        columns_names += "\"" + col + "\""

        if (idx + 1) < len(df.columns):
            columns_names += ","

    columns_names += "]"

    return columns_names
```
